# 一元回归分析
$$
    \hat y = \beta_{0} + \beta_{1}x + \epsilon
$$
我们假设
1. 所有自变量$x$的误差$\epsilon$互相独立
2. 所有自变量$x$的误差$\epsilon$服从均值为0，方差为$\sigma^2$的正态分布

## 统计量计算
### $\sigma^2$
$\sigma^2$的样本估计值为样本方差$s^2$
1. $s^2$ 直观上应该由SSE构造
$$
    \begin{array}{ll}
    s^2 = \frac{SSE}{(n-2)} \\
    则我们可以构造出一个卡方分布 \\
    \chi^2 = \frac{SSE}{\sigma^2} = \frac{(n-2)s^2}{\sigma^2} \\
    最后得出 \\
    E(s^2) = \sigma^2 \\
    \end{array}
$$
\* 即$s^2$是$\sigma^2$的无偏估计 <br>
2. 使用$s$能够确定拟合数据分布在哪个区间段中

### $\hat \beta$
1. $\hat \beta$ 是 $\beta$的无偏估计
2. $\hat \beta$ 的标准误差是
$$
    \sigma_{\hat \beta_{i}} = \sigma \cdot {{(X^TX)}^{-1}}_{ii}
$$
\* 其中${(X^TX)}^{-1}$是协方差矩阵的相应对角线位置值，也就是第$i$个变量的方差值

## 最小二乘法 LSM
注意，总体 $y$ 的样本平均值为 $\bar y$，预测值为 $\hat y$<br>
(Least Squared Method)
$$
    \hat y = \beta_{0} + \beta_{1}x
$$
使用MSE(Mean Squared Error)进行误差分析<br>
MSE的表现形式与SSE相同，即
$$
    SSE = \sum^{N}_{i}{{(y_{i}-\hat y_{i})}^2}
$$
这时，最小二乘估计公式为：
$$
\left\{
\begin{array}{ll}
\hat \beta_{1} = \frac{SS_{xy}}{SS_{xx}} \\
\hat \beta_{0} = \bar y - \hat \beta_{1} \bar x \\
SS_{xy} = \sum_{i}{x_{i}y_{i}}-\frac{(\sum_{i}{x_{i}})(\sum_{i}{y_{i}})}{n} \\
SS_{xx} = \sum_{i}{{x_{i}}^2} - \frac{{(\sum_{i}{x_{i}})}^2}{n}\\
\end{array}
\right .
$$

## 矩阵形式
参数说明：
$$
\begin{array}{ll}
    X ~ 自变量矩阵 \\
    Y ~ 因变量矩阵\\
    \hat \beta ~ OLS参数矩阵\\
\end{array}
$$
公式：
$$
\begin{array}{ll}
    \hat \beta = {(X^TX)}^{-1}X^TY \\
    SSE = Y^TY - {\hat \beta}^TX^TY \\
    s^2 = MSE = \frac{SSE}{n-(k+1)} \\
    * k为自变量数量
\end{array}
$$

## 最小二乘法参数分析
1. $\hat \beta_{1}$是$\beta_{1}$的无偏估计，即
$$
    E[\hat \beta_{1}] = \beta_{1}
$$
2. $\hat \beta_{1}$ 的方差为
$$
    V[\hat \beta_{1}] = \frac{\sigma^2}{SS_{xx}}
$$
 \* 其中，$\sigma$是随机变量$y$的标准差，是一个总体值<br>
3. 当总体方差$\sigma^2$未知时，我们可以通过样本方差$s^2$来进行估计
$$
\left \{
\begin{array}{ll}
s^2 = \frac{SSE}{n-2}\\
SSE = SS_{yy} - \hat \beta_{1} SS_{xy} \\
SS_{yy} = \sum_{i}{{y_{i}}^2} - \frac{{(\sum_{i}{y_{i}})}^2}{n}\\
SS_{xy} = \sum_{i}{x_{i}y_{i}}-\frac{(\sum_{i}{x_{i}})(\sum_{i}{y_{i}})}{n} \\
\end{array}
\right .
$$
4. 此时，$\hat \beta_{1}$方差的无偏估计为
$$
 V[\hat \beta_{1}] = \frac{s^2}{SS_{xx}}
$$
5. 构造的t-student检验统计量为
$$
    H_{0} : \beta_{1}=0 \\
    - 该值说明我们使用\hat y 进行线性估计时\beta_{1}的大小 \\
    - 一般使用\beta_{1}=0，即直接使用平均值\bar y进行拟合分析\\
    H_{1} : \beta_{1}\ne 0 \\
    T_{c} = \frac{\hat \beta_{1} - 0}{s /\sqrt{SS_{xx}}}
$$

## 相关系数
皮尔逊相关系数计算公式<br>
$r$的正负决定正相关与负相关；绝对值决定相关性的大小
$$
    r = \frac{SS_{xy}}{\sqrt{SS_{xx}SS_{yy}}}
$$

## 决定系数
如果使用平均值$\bar y$进行平行于x轴的线性估计，我们的残差平方和为
$$
    SS_{yy} = \sum_{i}{{(y_{i}-\bar y)}^2}
$$
如果使用预测值$\hat y$进行有斜率的线性估计，我们的离差平方和为
$$
    SSE = \sum_{i}{{(y_{i}-\hat y)}^2}
$$
决定系数$r^2$说明$y$关于预测值$\hat y$能够归因于$x$与$y$线性关系的比例，这个值越大说明线性关系贡献的比率越大
$$
    r^2 = 1 - \frac{SSE}{SS_{yy}}
$$
如果$r^2$较小，那么我们有理由认为$x$对于$y$的线性贡献值较小，此时应当考虑其他的自变量取值

## 利用模型进行估计和预测
一共分为两个方面：<br>首先是对于多次预测取平均值的预测即$E[y]$的预测；<br>其次是对于单个点$y$的预测<br>
在预测过程中，仍旧通过构造学生式t分布的方法进行置信区间的计算<br>

## 残差分析
残差分析即是对于$\epsilon$的分析，为了确保对于$\epsilon$的假设是正确的
1. 检验期望：绘制残差图$(y - \hat y) - x$
2. 检验方差：绘制残差图$(y - \hat y) - \hat y$
3. 检验正态分布： 绘制茎叶图或是直方图
4. 检验相互独立： 绘制残差-时间序列对照图

## 实验设计方差分析
### 逻辑
1. 对于不同处理分组中的均值进行是否相同的检验

### 单因子完全随机化设计
1. 一个处理就相当于是一个因子的一个水平
2. 该设计中只存在一个因子，并且该因子只有两个水平
3. 完全随机化是指将处理随机指派给试验单位

$$
    \begin{array}{ll}
    H_0 : \beta_1=\beta_2=\cdots=\beta_{p-1}=0 \\
    简化模型 ： E(y) = \beta_0 \\
    完全模型 ： E(y) = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_{p-1} x_{p-1} \\
    F=\frac{(SSE_R-SSE_C)/(p-1)}{SSE_C/(n-p)} \\
    * 其中，p是处理的数量，对应k+1 
    \end{array}
$$

### ANOVA分析
又名变异数分析，Analysis of Variance，主要探索连续型因变量与类别型自变量的关系

```python
def anova(data,y):
    '''
    data ~ input matrix
    y ~ output (n,) vector
    return ~ beta (n,) vector and SSE MSE value
    '''
    XTX = np.dot(data.T,data)
    XY = np.dot(data.T,y)
    beta = np.dot(np.linalg.inv(XTX),XY)
    YTY = np.dot(y.T,y)
    bXY = np.dot(beta.T,XY)
    SSE = YTY - bXY
    MSE = SSE/len(data.columns)
    return [beta, SSE, MSE]
```
