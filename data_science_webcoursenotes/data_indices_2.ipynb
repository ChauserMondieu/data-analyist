# 用户埋点
## 埋点与埋点表
### 通用业务（核心功能）
1. 基本不会变，长期可维护、可扩展
2. 需要考虑性能，分区分表分层等
3. 回归测试友好
### 临时业务（临时活动）
1. 临时活动，活动后功能会移除
2. 关注目标，只埋跟目标有关的即可
3. 没有爆发的情况下可以忽略性能

## 产品经理
### 分类
#### C端产品
服务于个人用户，通过广告引流获取价值<br>
需求分散，需要抽象与提炼<br>
功能简单，护城河较低，更关注用户体验，业务逻辑相对简单<br>
关注 转化率、留存率、传播率<br>
1. 获取用户
2. 留存用户
3. 流量变现

#### B端产品
服务于企业用户，为企业提供定制化服务<br>
客户需求明确，针对具体环节提升效率<br>
专业度相对较高，承包脏活累活，护城河较高，设计角色权限等相对繁琐<br>
关注 各个业务节点的成本和效率<br>
1. 把细分场景标准化
2. 提效降本
3. 服务变现

### 职责
1. 需求分析
    * 需求背景
2. 需求目标
3. 需求列表
4. 需求描述
5. 需求文档 + 原型图 + 流程图
6. 数据埋点+数据分析 -> 分析是否达到预期，根据结果调整迭代预期

## MVP[Minimum Viable Product]模型
1. 小步快包、快速迭代
2. 小成本投入，快速试错
3. 优先完成核心功能即刻上线接受检验

### 步骤
1. 梳理核心流程，找到核心功能点
2. 完成原型图绘制，产出原型demo
3. 产出需求文档、UI设计稿、需求评审
4. 研发、测试、上线
5. 计算投入产出比ROI
    * 转化链路是否存在问题
    * 业务模式是否存在问题

### 常见业务模式
媒介 + 内容、角色的链接
1. 决策形态型
    * 场景 - 输入 - 决策规则 - 输出
2. 流水线型
    * 内部协作的业务流
3. 售卖型
    * 对商品的定义、订购流程的设计
    * 根据商业模式设计实现
4. 人工服务型
