# $\beta$分布

$$
    f(y) = 
    \left\{
    \begin{array}{ll}
    \frac{ y^{\alpha-1}e^{-\frac{y}{\beta}}}{\beta^{\alpha}\Gamma(\alpha)} & ,0\le y < \infty \\
    0 & ,else
    \end{array}
    \right.
$$

```python
import scipy.stats as ss
import numpy as np
from matplotlib import pyplot as plt

x_beta = np.arange(-0.1,1.1,0.01)

plt.figure(figsize=(20,8),dpi=80)
plt.subplots_adjust(left=None,top=None,right=None,bottom=None,wspace=0.2, hspace=0.6)

for i in range(25):
    y_beta = ss.beta(int(i/5)+1, i%5+1).pdf(x_beta)
    plt.subplot(5,5,i+1).plot(x_beta, y_beta)
    plt.subplot(5,5,i+1).set_title(r'$alpha$=%.1f,$beta$=%.1f'%(int(i/5)+1,i%5+1))

plt.title(r'$\beta$ distribution with different $\alpha$ and $\beta$')
```
