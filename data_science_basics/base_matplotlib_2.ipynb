# 绘制散点图

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
import numpy as np
import random
import math
```

```python
my_font=font_manager.FontProperties(fname="/System/Library/Fonts/Hiragino Sans GB.ttc")

random.seed(0)
y_3=[random.randint(10,25) for i in range(1,32)]
random.seed(1)
y_10=[random.randint(5,25) for i in range(1,32)]

plt.figure(figsize=(20,8),dpi=80)

x_3=range(1,32)
x_10=range(51,82)
_x=list(x_3)+list(x_10)

_x_3_xtick_label=["3.{:02d}".format(i) for i in range(1,32)]
_x_10_xtick_label=["10.{:02d}".format(i-50) for i in range(51,82)]
_xtick_labels=_x_3_xtick_label+_x_10_xtick_label

plt.xticks(_x[::2],_xtick_labels[::2],rotation=45)
plt.yticks(range(min(y_3+y_10)-1,max(y_3+y_10)+1),rotation=45)

plt.xlabel("time",fontproperties=my_font)
plt.ylabel("temperature (unit:$^\circ$C)",fontproperties=my_font)
plt.title("temperature of March and October")

plt.scatter(x_3,y_3,label="March",color="red")
plt.scatter(x_10,y_10,label="October",color="blue")

plt.grid(alpha=0.2)
plt.legend(prop=my_font,loc="best")

plt.show()
```

# 直方图练习

```python
x=["猩球崛起3:终极之战","敦刻尔克","蜘蛛侠：英雄归来","战狼2"]
y_16=[15746,312,4497,319]
y_15=[12357,156,2045,168]
y_14=[2358,399,2358,362]

bar_width=0.2

plt.figure(figsize=(20,6),dpi=80)
my_font = font_manager.FontProperties(fname="/System/Library/Fonts/Hiragino Sans GB.ttc")

_x_14=list(range(len(x)))
_x_15=[i+bar_width for i in _x_14]
_x_16=[i+bar_width*2 for i in _x_14]
y=y_14+y_15+y_16
_y=range(min(y)-100, max(y)+100)

plt.xticks(_x_15,x,fontproperties=my_font)
plt.yticks(_y[::1000])
plt.title("movie rates from day 14 to day 16",fontproperties=my_font)

plt.xlabel("movie name",fontproperties=my_font)
plt.ylabel("rate")

plt.bar(_x_16,y_16,label="day16",color="magenta",width=bar_width)
plt.bar(_x_15,y_15,label="day15",color="orange",width=bar_width)
plt.bar(_x_14,y_14,label="day14",color="red",width=bar_width)

plt.grid(alpha=0.2)
plt.legend(prop=my_font,loc="best")

plt.show()
```
