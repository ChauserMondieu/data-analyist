```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
import numpy as np
import pandas as pd
```

```python
x = np.linspace(1,100,10000)
y = np.sin(x)

plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y,c='coral',lw=2,ls=':',marker='o',markersize=0.5)
```

```python
X = np.arange(0,1000,20).reshape([5,10])
for i in range(len(X[0,:])):
    X[1,i] *= 50

plt.figure(figsize=(12,8),dpi=80)
ax = plt.subplot(1,1,1, facecolor='whitesmoke')
sc = ax.scatter(np.array(X[0,:]),np.array(X[2,:]),
            s=np.array(X[1,:]),label='scatter',
            c=X[1,:],cmap=plt.get_cmap("jet"),
            alpha=0.9)
plt.legend(loc='best')
plt.colorbar(sc, ax=ax)
plt.title("this is a title")
```

# Seaborn

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1,7):
        plt.plot(x, np.sin(x + i * .5) * (7-i)*flip, label='sin%02d'%i)
    plt.legend()
sinplot()
```

```python
sns.set()
sinplot()
```

```python
import numpy as np 

np.random.seed(0)

def compute_reciprocols(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1,100,size=1000000)
%timeit compute_reciprocols(big_array)
```
