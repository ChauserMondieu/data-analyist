# basic
## figure axes axis
figure是 画布 <br>
axes是 图像 <br>
axis是 坐标轴 <br>
plt.gcf() 获取当前figure<br>
plt.gca() 获取当前axes<br>
## 绘图方式
第一种 <br>
plt.figure(num, figsize, dpi, facecolor, edgecolor)<br>
ax = plt.add_subplot(nrow,ncolumn,pos)<br>
ax.plot()<br>
第二种 <br>
fig,axes = plt.subplots(nrow,ncolumn)<br>
ax = axes[pos]<br>
ax.plot()<br>

```python
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import font_manager
import numpy as np
import random
```

```python
# font = {'family':'DejaVu Sans',
#         'weight': 'bold',
#         'size': 12}
# matplotlib.rc("font", **font)

# 通过font-manager引入字体库
# homebrew install fontconfig 
# fc-list :lang=zh 查看字体存储位置
my_font = font_manager.FontProperties(fname="/System/Library/Fonts/Hiragino Sans GB.ttc")

random.seed(0)
x = range(0,120)
y = [random.randint(20,35) for i in range(120)]
plt.figure(figsize=(10,5),dpi=80)

plt.plot(x,y)

# 设置x轴y轴标签
_x=list(x)
_xtick_labels = ["10:{:02d}".format(i) for i in range(60)]
_xtick_labels += ["11:{:02d}".format(i) for i in range(60)]

# 数值与字符串一一对应
plt.xticks(_x[::5],_xtick_labels[::5],rotation=45)
plt.yticks(range(20,36))


# 设置x轴标签
plt.xlabel("time",fontproperties=my_font)
plt.ylabel("temperature (unit:$^\circ$C)",fontproperties=my_font)
plt.title("10:00 -12:00 temperature change in every minute",fontproperties=my_font)

plt.show()
```

```python
x = range(11,31)
y1 = [1,0,1,1,2,4,3,2,3,4,4,5,6,5,4,3,3,1,1,1]
y2 = [1,1,2,3,4,5,4,4,3,1,3,2,3,4,2,1,1,1,1,1]

plt.figure(figsize=(10,5),dpi=80)
my_font=font_manager.FontProperties(fname="/System/Library/Fonts/Hiragino Sans GB.ttc")

# color linestyle linewidth alpha
plt.plot(x,y1,label=r'$\alpha$',color="orange",linestyle=":")
plt.plot(x,y2,label=r"$\beta$",color="cyan",linestyle="-.")

_x =list(x)
_xtick_labels= ["age{}".format(i) for i in _x]
plt.xticks(_x[::2],_xtick_labels[::2],rotation=45)
plt.yticks(list(range(min(y),max(y)*2)))

plt.xlabel("年龄",fontproperties=my_font)
plt.ylabel("数量",fontproperties=my_font)
plt.title("男｜女朋友数量",fontproperties=my_font)

plt.grid(alpha=0.4)

plt.legend(prop=my_font, loc="best")

plt.show()
```
